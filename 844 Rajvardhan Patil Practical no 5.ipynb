```python

import matplotlib.pyplot as plt
fo=open('testmarks1.csv')
data=fo.read()
listline=data.splitlines()
x_rno=[]
y_edsmarks=[]
for line in listline:
    word=line.split(",")
    if(word[0].isdigit() or word[1].isdigit()):
        x_rno.append(word[0])
        y_edsmarks.append(float(word[1]))
        
print(x_rno)
print(y_edsmarks)
plt.title("Result analysis of EDS")
plt.xlabel("ROLL NO")
plt.ylabel("MARKS")
plt.plot(x_rno,y_edsmarks)
plt.show()
```
